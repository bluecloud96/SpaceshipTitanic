# 데이터 피쳐링

## PassengerId, Name, Cabin

#### PassengerId의 경우 gggg_pp 형식으로 gggg가 같으면 같은 그룹이고 가족인 경우도 있고 아닌 경우도 있다.

```python
df = pd.concat([train_df, test_df])
```

```python
df['GroupId'] = df['PassengerId'].apply(lambda x: x.split('_')[0])
```

```python
df['GroupId'].describe(include = 'all')
```

#### group이 너무 많아서 group만으로는 추리기 힘듬 groupsize로 추리기

```python
df['GroupSize'] = df['GroupId'].map(df['GroupId'].value_counts())
```

```python
df['GroupSize'].describe()
```

```python
train_df = df.iloc[:8693]
```

```python
sns.countplot(data=train_df, x='GroupSize',hue = "Transported")
```

```python
df['is_alone'] = df['GroupSize'] == 1
```

```python
df
```

```python
train_df = df.iloc[:8693]
```

```python
sns.countplot(data=train_df, x='is_alone',hue = "Transported")
```

```python
df
```

#### cabin의 경우도 객실추리기

```python
import numpy as np
```

```python
df['Cabin_deck'] = df['Cabin'].apply(lambda x: x.split('/')[0] if pd.notnull(x) else np.nan)
df['Cabin_side'] = df['Cabin'].apply(lambda x: x.split('/')[2] if pd.notnull(x) else np.nan)
```

```python
train_df = df.iloc[:8693]
```

```python
order = train_df['Cabin_deck'].value_counts().index
sns.countplot(data=train_df, x='Cabin_deck',hue = "Transported", order = order)
```

#### cabin_deck은 제일 빈도수가 높은 f, g빼고 나머지로 나누었다.

```python
df['Cabin_deck_grouped'] = df['Cabin_deck'].apply(lambda x: x if pd.notna(x) and x in ['F', 'G'] else ('Other' if pd.notna(x) else np.nan))
```

```python
train_df = df.iloc[:8693]
```

```python
order = train_df['Cabin_deck_grouped'].value_counts().index
sns.countplot(data=train_df, x='Cabin_deck_grouped',hue = "Transported", order = order)
```

```python
df['Cabin_num'] = df['Cabin'].apply(lambda x: int(x.split('/')[1]) if pd.notnull(x) else np.nan)
```

```python
df['Cabin_num'].describe()
```

```python
train_df = df[:8693]
```

```python
plt.figure(figsize=(10, 6))
plt.title("Cabin_num and Transported")
sns.violinplot(
    x = "Transported",
    y = "Cabin_num",
    data =train_df)
```

```python
sns.countplot(data=train_df, x='Cabin_side',hue = "Transported")
```
